```python
# import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjust the configuration of the plots we will create

# read in the data

df=pd.read_csv(r"C:\Users\hp\Downloads\movies.csv")
```

```python
# let's look at data 
df.head()
```

```python
# let's see if there is missing data

for col in df.columns:
    pc_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pc_missing))
```

```python
# data types for my columns

df.dtypes
```

```python
# change the data type of columns 'budget' and 'gross'

df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['yearcorrected']= df['released'].astype(str).str[:4]
df.head()
```

```python
df.sort_values(['gross'], inplace = False, ascending = False).head()
```

```python
# displaying all rows
pd.set_option('display.max_rows', None)
```

```python
# droping duplicates
df.drop_duplicates().head()
```

```python
 df.head()
```

```python
# scatter plot with budget vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel(' Budget For Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
df.dtypes
```

```python
# let's look at correlation
df.corr(numeric_only=True,method='pearson') # pearson, kendall, spearman
```

```python
# high correlation between budget and gross
```

```python
correlation_matrix=df.corr(numeric_only=True,method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel(' Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerize=df

for col_name in df_numerize.columns:
    if(df_numerize[col_name].dtype == 'object'):
        df_numerize[col_name]=df_numerize[col_name].astype('category')
        df_numerize[col_name]=df_numerize[col_name].cat.codes

df_numerize.sort_values(['gross'], inplace = False, ascending = False).head()
```

```python
df.sort_values(['gross'], inplace = False, ascending = False).head()
```

```python
correlation_matrix=df_numerize.corr(numeric_only=True,method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matric for Numeric Features')
plt.xlabel(' Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
